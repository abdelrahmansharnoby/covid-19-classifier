```python
import pandas as pd 
import numpy as np 
import math

from sklearn.linear_model import LogisticRegression 
from sklearn.preprocessing import StandardScaler, MinMaxScaler, OrdinalEncoder, LabelEncoder, OrdinalEncoder
from sklearn.compose import make_column_transformer
from sklearn.model_selection import train_test_split

from sklearn.metrics import recall_score, f1_score, precision_score, accuracy_score, classification_report,\
roc_auc_score, confusion_matrix

import matplotlib.pyplot as plt
```

```python
# reading the data into a list 
dataframe = pd.read_csv("data.csv")
print ( len(dataframe) )
dataframe.head() 
```

```python
lb = LabelEncoder()
for i in dataframe.columns:
    dataframe[i]=lb.fit_transform(dataframe[i])
dataframe
```

```python
#splitting the data into Training (80%) and Testing (20%)
x = dataframe.iloc[:, 1:13]
y = dataframe["result"]
xTrain , xTest , yTrain , yTest = train_test_split(x,y,random_state=2,test_size=0.2,shuffle = True)
xTrain
#pd.to_numeric(y)
#print (type(y))
```

```python
#transformer = Normalizer().fit(xTrain)
#xTrain = transformer.transform(xTrain)
#xTest = transformer.transform(xTest)
```

```python
reg = LogisticRegression (max_iter=5000)
reg.fit(xTrain , yTrain)
```

```python
yPred = reg.predict(xTest)
yPred
```

```python
print (accuracy_score(yTest,yPred))
```

```python
print (f1_score(yTest,yPred))
```

```python
print (recall_score(yTest,yPred))
```

```python
print (precision_score(yTest,yPred))
```

```python
print (roc_auc_score(yTest,yPred))
```

```python
print (confusion_matrix(yTest,yPred))
```
